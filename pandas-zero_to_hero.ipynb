## **Background and Motivation**

Pandas is a powerful and easy to use data analysis tool built for the Python ecosystem. This software was originally constructed at a capital management firm out of the necessity for a fast and flexible tool for performing quantitative analysis on financial data. 

Some of the features include:
  * reading and writing of data between in-memory dataframes and various file formats
  * data alignment and easy handling of missing data
  * reshaping and pivoting of datasets
  * slicing, indexing and subsetting datasets
  * insertion and deletion
  * group by engine allowing various aggregations
  * merging and joining
  * time series functionality such as date ranges, frequency conversions, moving window statistics, date shifting and lagging
  * data filtration
